```python
def get_item_url():
    url = 'https://www.ebay.com/sch/i.html?_nkw=playstation+4+slim&LH_BIN=1&rt=nc&_ipg=100'
    user_agent = {'User-agent': 'Mozilla/5.0'}
    sponsored_list = []
    non_sponsored_list = []
    for pgn in range(1, 11): #
        url_pgn = url + '&_pgn=' + str(pgn)#the first 10 search pages
        response = requests.get(url_pgn, headers = user_agent)
        content=response.text#get content from search pages in form of text
        a_list=re.findall(re.compile(r'(?<=class="s-item__link").+?(?=a>)'),str(content))#regex to get search result
        for a in a_list:
            href = re.findall(re.compile(r'(?<=href=").+?(?=")'), str(a))#regex to extract link of items
            if "S" in a and "P" in a and "O" in a and "N" in a and "S" in a and "O" in a and "R" in a and "E" in a and "D" in a:#items sponsored
                sponsored_list.extend(href)
            else:
                non_sponsored_list.extend(href)
        time.sleep(5)
    print(len(sponsored_list),"\n",len(non_sponsored_list))
    with open("sponsored.txt","w") as f:#write result into a text file
        for sl in sponsored_list:
            f.write(sl+"\n")
    with open("non_sponsored.txt","w") as f:
        for nsl in non_sponsored_list:
            f.write(nsl+"\n")

def down_load_htm():
    user_agent = {'User-agent': 'Mozilla/5.0'}
    with open("sponsored.txt", "r") as f:#read urls
        sponsored_list=f.readlines()
    with open("non_sponsored.txt", "r") as f:
        non_sponsored_list=f.readlines()

    htm_num=len(os.listdir("sponsored/"))#get the number of files in folder "sponsored"
    print("htm_num",htm_num)
    for i in range(htm_num,len(sponsored_list)):#iterate i from the number of files to Total number of urls
        #number of files = number of urls already get
        sl=sponsored_list[i]
        id=(sl.split("?")[0]).split("/")[-1]#item id from url
        filename = id+'.htm'
        try:
            response = requests.get(sl, headers=user_agent)#request item url
            rt=response.text#get page content
        except:
            print("error occured")
            rt="error occured"
        with open("sponsored/"+filename, 'w',encoding='gb18030') as f:
            f.write(rt)
        print("filename:",filename)
        #input("-")
        time.sleep(5)

    print("sponsored finished,starting non_sponsored")
    non_htm_num = len(os.listdir("non_sponsored/"))#repeat what has been done in "sponsored"
    print("non_htm_num", non_htm_num)
    for i in range(non_htm_num, len(non_sponsored_list)):
        nsl = non_sponsored_list[i]
        id=(nsl.split("?")[0]).split("/")[-1]
        filename = id+'.htm'
        try:
            response = requests.get(nsl, headers=user_agent)
            rt=response.text
        except:
            print("error occured")
            rt="error occured"
        with open("non_sponsored/"+filename, 'w',encoding='gb18030') as f:
            f.write(rt)
        print("filename:", filename)
        time.sleep(5)

def get_htm_data():
    for s in ["sponsored/","non_sponsored/"]:
        for f in os.listdir(s):
            #f = "283771194867.htm"#best offer
            #f = "361874940467.htm"#no best offer
            #f="324037896258.htm"
            print("html_name",f)
        if f.endswith("htm"):
            with open(s+f,"r+",encoding='gb18030') as c:
                content=c.read()#read page content
                #print(content)

                sellers=(re.findall(re.compile(r'(?<=aria-label="Member ID:).+?(?=")'),str(content))[0]).replace("&nbsp;","")
                print("sellers:",sellers) #seller name

                try:
                    score=re.findall(re.compile(r'(?<=alt="feedback score:).+?(?=")'),str(content))[0] #seller score
                except:
                    score="-"
                print("score:", score)

                price = re.findall(re.compile(r'(?<=id="prcIsum").+?(?=</span>)'), str(content)) #item price
                if price!=[]: 
                    price=price[0].split(">")[-1]
                else:
                    try:
                        price = re.findall(re.compile(r'(?<=id="mm-saleDscPrc").+?(?=</span>)'), str(content))[0]
                        price=price.split(">")[-1]
                    except:
                        price="-"
                print("price:", price)

                sold = re.findall(re.compile(r'(?<=vi-qtyS-hot-red  vi-qty-vert-algn vi-qty-pur-lnk).+?(?=</a>)', re.S), str(content))#number of sold
                if sold!=[]:
                    sold=re.sub('\\<.*?\\>','',sold[0]).strip()
                else:
                    sold="-"
                print("sold:", sold)

                best_offer = re.findall(re.compile(r'(?<=d="boBtn_btn").+?(?=</a>)',re.S), str(content))#best offer available
                if best_offer!=[]:
                    best_offer="Yes"
                else:
                    best_offer="No"
                print("best_offer:",best_offer)

                title = re.findall(re.compile(r'(?<=<title>).+?(?=</title>)'), str(content))[0] #item title
                title=title.split("|")[0]
                print("title:", title) 


                if "No returns" in str(content):#returns or no returns
                    if_return="No returns"
                elif "Returns accepted" in str(content) and "No returns" not in str(content):
                    if_return="Returns accepted"
                else:
                    if_return="-"
                print("if_return:",if_return)

                try:
                    shipping_price = re.findall(re.compile(r'(?<=id="fshippingCost").+?(?=</span>)',re.S), str(content))[0] #shipping price
                    shipping_price=shipping_price.split(">")[-1]
                except:
                    shipping_price="-"
                print("shipping_price:", shipping_price)
                
                try:
                    condition = re.findall(re.compile(r'(?<=id="vi-itm-cond").+?(?=</div>)',re.S), str(content))[0] #item's condition
                    condition = condition.split(">")[-1]
                except:
                    condition = "-"
                print("condition:", condition,"\n\n") 

if __name__ == "__main__":
    #prior set up folder "sponsored", foler "non_sponsored"
    get_item_url()#get list of item urls
    down_load_htm()
    get_htm_data()
```
