```python
"""
回答问题一：
可以看出，Meta评分比imdb自家评分普遍稍低一些，meta评分的std更大意味着分数分布更分散。

究其原因，因为Meta用户为专业及近专业影评人，相比imdb用户为业余影迷，meta用户对影片有更严格的要求和评判。
"""
#问题二：2006年到2016年间票房变化的趋势如何？
yrbox = submovie_df.groupby('年份')['票房'].sum()
yrbox
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

fig = plt.figure()
ax= fig.add_subplot()
sns.catplot(x='年份', y='票房', kind = 'box', data=submovie_df)
```

```python
'''
回答问题二：
票房总量是逐年增加的，2016年有激增。
但单独看到每一年的票房，中位数变动并不大。
我们可以说，票房的市场整体并没有升温，但在过去几年尤其2016年，越来越多的电影产生了单部2亿美元以上的票房。
'''
#问题三：2006-2016年间，上榜imdb的次数最多的导演是哪5个人？上榜imdb的评分最高的导演是哪5个人？导演们上榜的次数和影片的评分，有没有关系？
dir_counts = submovie_df['导演'].value_counts()
dir_counts.head()
```

```python
dir_counts.name='导演片数'
```

```python
dir_ratings=submovie_df.groupby('导演')['评分'].mean().sort_values(ascending=False)
dir_ratings.head()
```

```python
frame=(dir_counts,dir_ratings)
answer3=pd.concat(frame,axis=1, sort=True)
answer3
```

```python
answer3.sort_values("评分",inplace=True, ascending= False)
print(answer3)
```

```python
fig = plt.figure(figsize=(12,28))
ax= fig.add_subplot()
sns.catplot(x="导演片数", y="评分", kind="bar", data=answer3, ax=ax)
```

```python
'''
回答问题三：
2006-2016年间，
上榜次数最多的导演是：Ridley Scott, Paul W.S. Anderson, David Yates, Michael Bay, Peter Berg
作品评分最高的导演是：Christopher Nolan, Olivier Nakache, Makoto Shinkai, Aamir Khan, Florian Henckel von Donnersmarck 
两项排名没有导演同时进入前五名。
由上榜次数和评分的可视化结果，我们可以看出，导演们上榜的次数和他们作品的评分，没有明显关系。
'''
```
