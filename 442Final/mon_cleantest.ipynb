```python
for k in range(len(mond)):
    value=mond["Salary"][k]
    if(type(value)==str):
        if "hour" in value:
            if "," in value:
                mond["Salary"].values[k]=value.replace("hour","year")
        if 'hour' in value:
            new_num=[]
            value=value.replace(",", "")
            num=re.findall(r'\d+',value)
            for i in num:
                if float(i)!=0:
                    new_num.append(float(i))
            
            
            new_num=sum([int(i) for i in new_num])/len(new_num)
            mond["Salary"].values[k]=new_num*52*40
            #print(num)
        elif 'week' in value:
            print(value)
            new_num=[]
            value=value.replace(",", "")
            num=re.findall(r'\d+',value)
            for i in num:
                if float(i)!=0:
                    new_num.append(float(i))
            
            
            new_num=sum([int(i) for i in new_num])/len(new_num)
            
            mond["Salary"].values[k]=new_num*52
        elif 'month' in value:
            print(value)
            new_num=[]
            value=value.replace(",", "")
            num=re.findall(r'\d+',value)
            for i in num:
                if float(i)!=0:
                    new_num.append(float(i))
            
            
            new_num=sum([int(i) for i in new_num])/len(new_num)
            
            mond["Salary"].values[k]=new_num*12
        elif 'year' in value:
            new_num=[]
            value=value.replace(",", "")
            num=re.findall(r'\d+',value)
            for i in num:
                if float(i)!=0:
                    new_num.append(float(i))
            
            
            new_num=sum([int(i) for i in new_num])/len(new_num)
            
        #print(num)
            mond["Salary"].values[k]=new_num
        else:
            mond["Salary"].values[k]=0
```
