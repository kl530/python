```python

from bs4 import BeautifulSoup
import numpy as np
import pandas as pd

def walkFile(file):
    import os
    filenames = []
    for root, dirs, files in os.walk(file):
        for f in files:
            name = os.path.join(root, f)
            filenames.append(name)
    return filenames

def parse(name):
    html = open(name, encoding='utf-8').read()
    soup = BeautifulSoup(html, 'html.parser')
    title = soup.find_all('div', class_='jobsearch-JobInfoHeader-title-container')[0].find_all('h3')[0].text
    try:
        company = soup.find_all('div', class_='icl-u-lg-mr--sm icl-u-xs-mr--xs')[0].find_all('a')[0].text
    except:
        company = ''
    try:
        rating = soup.find_all('div', class_='icl-Ratings icl-Ratings--gold icl-Ratings--sm')[0].find_all('a')[0].get('aria-label')
        rating = rating.split()[0]
    except:
        rating = ''
    try:
        review = soup.find_all('div', class_='icl-Ratings-count')[0].text
    except:
        review = ''
    try:
        location = soup.find_all('div', class_='jobsearch-InlineCompanyRating icl-u-xs-mt--xs jobsearch-DesktopStickyContainer-companyrating')[0].find_all('div')[-1].text
    except:
        location = ''
    d = {}
    d['title'] = title
    d['company'] = company
    d['rating'] = rating
    d['review'] = review
    d['location'] = location
    return d

df = pd.DataFrame(columns=('title', 'company', 'rating', 'review', 'location'))

filenames = walkFile('./entry')
for name in filenames:
    if 'content' in name:
        d = parse(name)
        df = df.append(d, ignore_index=True)

print(df)

```
